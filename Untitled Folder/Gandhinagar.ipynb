```python
import requests

from bs4 import BeautifulSoup

from datetime import datetime
from dateutil import parser
```

```python
urls = {
    "icu w/": "https://vmc.gov.in/HospitalModuleGMC/HospitalBedsDetails.aspx?tid=2",
    "icu w/o": "https://vmc.gov.in/HospitalModuleGMC/HospitalBedsDetails.aspx?tid=3",
    "bed w/": "https://vmc.gov.in/HospitalModuleGMC/HospitalBedsDetails.aspx?tid=4",
    "bed w/o": "https://vmc.gov.in/HospitalModuleGMC/HospitalBedsDetails.aspx?tid=5"
}
```

```python
r_type = "icu w/"
url = urls[r_type]
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
tableEl = soup.find("table")
tableEl
```

```python
rowNames = [th.get_text() for th in tableEl.find("tr").find_all("th")]
rowNames
```

```python
hospitals = {}

def parseHospital(tableRowEl):
    tds = tableRowEl.find_all("td")
    nameEl = tds[0]
    lastUpdated = tds[7].get_text()
    address = tds[1].get_text()
    total = int(tds[2].get_text())
    available = int(tds[4].get_text())
    print (nameEl.get_text(), lastUpdated, available, total, address)
    
for hospitalEl in tableEl.find_all("tr")[1:]:
    parseHospital(hospitalEl)
```
